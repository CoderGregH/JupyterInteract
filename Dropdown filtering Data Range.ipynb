```python
import ipywidgets as widgets
from ipywidgets import Output
from IPython.display import display
from ipywidgets import interact

out = Output()
```

# Capture Dataframe now its loaded


```python
with out:
    display(weather)
```

```python
ddl_month = widgets.Dropdown(options=weather_months)
sl_sunHours = widgets.IntSlider(min=10, step=10,max=300)
float_min_hilite = widgets.IntText(value=40)
```

```python
def hiltie_precip(precip):
    return 'color: {0}'.format('blue' if precip > float_min_hilite.value else 'black')
    #return 'color: {0}'.format('blue')
        
def get_filter(month,sun_hours):
    filteredByMonth = weather[weather.month == month]
    filteredByMonthAndSunHours = filteredByMonth[filteredByMonth.sun_hours >= sun_hours]
    #return filteredByMonthAndSunHours.style.applymap(hiltie_precip, subset=weather.columns[8:])
    return filteredByMonthAndSunHours
                                                 
```

```python
def ddl_month_observer(bunch):
    out.clear_output()
    with out:
        #display(weather[weather.month == bunch.new])
        display(get_filter(bunch.new,sl_sunHours.value))
        
def sl_sunHours_observer(bunch):
    out.clear_output()
    with out:
        #display(weather[weather.sun_hours > bunch.new])
        display(get_filter(ddl_month.value,bunch.new))
        
        
def float_min_hilite_observer(bunch):
    out.clear_output()
    with out:
        #display(weather[weather.sun_hours > bunch.new])
        display(get_filter(ddl_month.value,bunch.new))
```

```python
ddl_month.observe(ddl_month_observer, names='value')
sl_sunHours.observe(sl_sunHours_observer, names='value')
float_min_hilite.observe(float_min_hilite_observer,names='value')
```
